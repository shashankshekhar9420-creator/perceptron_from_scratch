```python
import random
import matplotlib.pyplot as plt
```

```python
def train_and_plot(X, y):
  weight=[random.random(),random.random()]
  bias=random.random()
  learning_rate=0.1

  errors=[]
  def predict(x,weight,bias):
    z=x[0]*weight[0]+x[1]*weight[1]+bias
    if z>0:
      return 1
    else:
      return 0

  for epoch in range(100):

    total_error_per_epoch=0

    for x in X:
      prediction=predict(x,weight,bias)
      error=y[X.index(x)]-prediction
      weight[0]=weight[0]+learning_rate*error*x[0]
      weight[1]=weight[1]+learning_rate*error*x[1]
      bias=bias+learning_rate*error

      total_error_per_epoch+=abs(error)

    errors.append((total_error_per_epoch))
  plt.plot(errors)
  plt.xlabel("Epoch")
  plt.ylabel("Total Error")
  plt.show()

  for x in X:
    if y[X.index(x)]==1:
      plt.scatter(x[0], x[1], marker='o', color='green')
    else:
      plt.scatter(x[0], x[1], marker='x', color='red')

  x1_vals = [0, 1]
  x2_vals = [-(weight[0]*x1 + bias) / weight[1] for x1 in x1_vals]
  plt.plot(x1_vals, x2_vals)
  plt.show()
```

```python
X = [(0,0),(0,1),(1,0),(1,1)]
```

```python
train_and_plot(X, [0,0,0,1])  # AND
```

```python
train_and_plot(X, [0,1,1,1])  # OR
```

```python
train_and_plot(X, [1,1,1,0])  # NAND
```

```python
train_and_plot(X, [0,1,1,0])  # XOR
```
